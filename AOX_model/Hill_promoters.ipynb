```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#experimental data Mrx1
exp_conc = [75, 110, 180, 200, 225, 300, 360, 410, 480]
exp_perc = [20, 32, 45, 50, 55, 61, 76, 80, 78]
plt.plot(exp_conc, exp_perc)
```

```python
# Hill function
mrx1 = np.linspace(75, 500)
Vmax = 100; kd = 200; n = 1
y = lambda mrx1, n: Vmax* mrx1**n/(kd**n + mrx1**n)
plt.plot(exp_conc, exp_perc, mrx1, y(mrx1, n))
```

```python
# Hill function
Vmax = 100; kd = 200; # define parameters
hill_perc = lambda mrx1, n: Vmax* mrx1**n/(kd**n + mrx1**n) # define Hill function
n_vec = np.arange(1,2, 0.001); min_mse = 100 # optimzation of parameter n fitting data
for n in n_vec: 
    mse = mean_squared_error(exp_perc, hill_perc(exp_conc, n))
    if mse < min_mse:
        n_opt = n
        min_mse = mse
        
print("Optimal Hill coefficient (n) : {:.3f}".format(n_opt))

mrx1 = np.linspace(75, 500)
plt.plot(exp_conc, exp_perc, mrx1, y(mrx1, n_opt))
```
